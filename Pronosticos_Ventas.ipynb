```python
import pandas as pd
```

```python
tesla = pd.read_csv("ventas_modelos.csv")
```

```python
tesla
```

```python
from sklearn import linear_model
```

```python
tesla["back"]  =tesla["Year"] - tesla["Salida"]
tesla
```

```python
import matplotlib.pyplot as plt
tesla["back"]  = tesla["Precio"]/(tesla["Year"] - tesla["Salida"] + 1)
X_multiple = tesla.filter(["Year", "Precio", "Salida", "back"])
y_multiple = tesla.filter(["Total"])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_multiple, y_multiple, test_size=0.2, random_state=42)
```

```python
lr_multiple = linear_model.LinearRegression()
```

```python
lr_multiple.fit(X_train, y_train)
```

```python
Y_pred_multiple = lr_multiple.predict(X_test)
```

```python
print('DATOS DEL MODELO REGRESIÓN LINEAL MULTIPLE')
print()
print('Valor de las pendientes o coeficientes "a":')
print(lr_multiple.coef_)
print('Valor de la intersección o coeficiente "b":')
print(lr_multiple.intercept_)
```

```python
print('Precisión del modelo:')
print(lr_multiple.score(X_train, y_train))
```

```python
precio = 25000
salida = 2025
years  = [2025, 2026, 2027, 2028, 2029, 2030]
predictions = {}
predictions["Hachback"] = []
predictions["year"] = years
for year in years:
    data = {"Year":[year], "price":[precio], "Salida":[salida], "back":[precio/(year - salida + 1)]}
    data  = pd.DataFrame(data=data)
    predictions["Hachback"].append(lr_multiple.predict(data)[0][0])
```

```python
years  = [2022, 2023, 2024, 2025, 2026, 2027, 2028, 2029, 2030]

models = tesla.Model.unique()
year = 2025
predictions_models = {}

for model in models:
    data_model = tesla[tesla["Model"] == model]
    predictions_models[model] = []
    for year in years:
        data = {"Year":[year], "price":[data_model.Precio.max()], "Salida":[data_model.Salida.max()], "back":[data_model.Precio.max()/(year - data_model.Salida.max() + 1)]}
        data  = pd.DataFrame(data=data)
        predictions_models[model].append(lr_multiple.predict(data)[0][0])
predictions_models
    
```
